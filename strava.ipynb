```python
import re
import requests
from stravalib import Client

# --- USER INPUT ---
MY_TOKEN = "d8a0e9e7389cd0e0d14b5900739e1b158d0578de"
MY_URL = "https://www.strava.com/activities/17364130694" 
# ------------------

def get_activity_data(url, token):
    # 1. Unfold mobile links to find the real ID
    if "strava.app.link" in url:
        print(" Unfolding mobile link...")
        url = requests.get(url, allow_redirects=True).url
    
    match = re.search(r"activities/(\d+)", url)
    if not match:
        raise ValueError("Could not find Activity ID. Try pasting the long URL from your browser.")
    
    activity_id = match.group(1)
    client = Client(access_token=token)
    
    # 2. Check what data Strava is willing to share
    print(f" Checking Activity: {activity_id}")
    streams = client.get_activity_streams(activity_id, types=["latlng", "altitude"], resolution="high")
    
    if not streams or "latlng" not in streams:
        print(" ERROR: No GPS data found.")
        print("FIX: Ensure activity privacy is 'Followers' or 'Everyone' and it's not an 'Indoor' workout.")
        return None, None
        
    return streams["latlng"].data, streams["altitude"].data

try:
    coords, alts = get_activity_data(MY_URL, MY_TOKEN)
    if coords:
        print(f" SUCCESS! Captured {len(coords)} GPS points.")
        print("Ready for Phase 3: The 2D Route Animation!")
except Exception as e:
    print(f" Unexpected Error: {e}")
```

```python
import pandas as pd
import plotly.graph_objects as go
import warnings

# Silence the warnings so they don't clutter your screen
warnings.filterwarnings("ignore", category=DeprecationWarning)

# 1. Prepare data (coords from image_4d70b0.png)
df = pd.DataFrame(coords, columns=['lat', 'lon'])

# 2. Use the NEW 'Scattermap' instead of 'Scattermapbox'
fig = go.Figure(
    data=[go.Scattermap(
        lat=df['lat'],
        lon=df['lon'],
        mode='lines',
        line=dict(width=4, color='#FC4C02'),
        name="Activity Path"
    )],
    layout=go.Layout(
        # 'open-street-map' is the most reliable free background
        map_style="open-street-map",
        map_zoom=13,
        map_center={"lat": df['lat'].mean(), "lon": df['lon'].mean()},
        title="Strava Activity Flyover - 2D Preview",
        height=600,
        updatemenus=[dict(
            type="buttons",
            buttons=[dict(label=" Play Animation", method="animate", args=[None])]
        )]
    ),
    frames=[go.Frame(
        data=[go.Scattermap(
            lat=df['lat'][:i], 
            lon=df['lon'][:i]
        )]
    ) for i in range(1, len(df), 30)] # Speed up animation by jumping 30 points
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# 1. Prepare the data
df = pd.DataFrame(coords, columns=['lat', 'lon'])
df['alt'] = alts

# 2. Create the 3D Figure
fig_3d = go.Figure()

# Add the route as a thick 3D line
fig_3d.add_trace(go.Scatter3d(
    x=df['lon'],
    y=df['lat'],
    z=df['alt'],
    mode='lines',
    line=dict(color='#FC4C02', width=8),
    name="Your Route"
))

# 3. Configure the 3D 'World'
fig_3d.update_layout(
    title="3D Cinematic Flyover (No-Token Version)",
    scene=dict(
        # We exaggerate the Z-axis (altitude) so your 20m hills look like mountains
        aspectratio=dict(x=1, y=1, z=0.4), 
        xaxis_title="Longitude",
        yaxis_title="Latitude",
        zaxis_title="Elevation (m)",
        bgcolor="rgb(20, 20, 20)", # Dark cinematic background
        camera=dict(
            eye=dict(x=1.2, y=1.2, z=0.5) # Drone-style follow angle
        )
    ),
    margin=dict(l=0, r=0, b=0, t=40),
    template="plotly_dark"
)

fig_3d.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# 1. Create the base 3D line
fig_flyover = go.Figure(data=[go.Scatter3d(
    x=df['lon'], y=df['lat'], z=df['alt'],
    mode='lines',
    line=dict(color='#FC4C02', width=8)
)])

# 2. Build the "Frames" for the camera movement
frames = []
# We'll create a frame every 50 points to keep the animation smooth but fast
step = 50 
for i in range(0, len(df), step):
    # Current point coordinates
    curr_x = df['lon'].iloc[i]
    curr_y = df['lat'].iloc[i]
    curr_z = df['alt'].iloc[i]
    
    frames.append(go.Frame(
        layout=dict(
            scene_camera=dict(
                # The 'eye' is the drone position relative to the center
                eye=dict(x=1.2, y=1.2, z=0.6),
                # The 'center' is where the camera is looking
                center=dict(x=0, y=0, z=0)
            ),
            # This centers the map on the current GPS point as we move
            scene=dict(
                xaxis=dict(range=[df['lon'].min(), df['lon'].max()]),
                yaxis=dict(range=[df['lat'].min(), df['lat'].max()]),
                zaxis=dict(range=[df['alt'].min() - 5, df['alt'].max() + 5])
            )
        ),
        name=f"frame{i}"
    ))

# 3. Add the Play Button and Final Layout
fig_flyover.update_layout(
    title="Final Phase: Cinematic 3D Flyover",
    template="plotly_dark",
    updatemenus=[dict(
        type="buttons",
        buttons=[dict(label=" Start Flyover", method="animate", args=[None, {"frame": {"duration": 100, "redraw": True}}])]
    )],
    scene=dict(
        aspectratio=dict(x=1, y=1, z=0.4),
        xaxis_title="Lon", yaxis_title="Lat", zaxis_title="Alt"
    )
)

fig_flyover.frames = frames
fig_flyover.show()
```
